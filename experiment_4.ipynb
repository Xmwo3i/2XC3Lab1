```python
import time
import random
import matplotlib.pyplot as plt

from good_sorts import quicksort, mergesort, heapsort
```

```python
def create_random_list(n):
    return [random.randint(0, n) for _ in range(n)]
```

```python
sizes = [500, 1000, 2000, 4000, 8000]
runs = 5

quick_times = []
merge_times = []
heap_times = []
```

```python
for n in sizes:
    q_total = m_total = h_total = 0

    for _ in range(runs):
        L = create_random_list(n)

        A = L.copy()
        start = time.time()
        quicksort(A)
        q_total += time.time() - start

        A = L.copy()
        start = time.time()
        mergesort(A)
        m_total += time.time() - start

        A = L.copy()
        start = time.time()
        heapsort(A)
        h_total += time.time() - start

    quick_times.append(q_total / runs)
    merge_times.append(m_total / runs)
    heap_times.append(h_total / runs)
```

```python
plt.figure()
plt.plot(sizes, quick_times, label="Quick Sort")
plt.plot(sizes, merge_times, label="Merge Sort")
plt.plot(sizes, heap_times, label="Heap Sort")

plt.xlabel("List length")
plt.ylabel("Time (seconds)")
plt.legend()
plt.savefig("experiment4_graph.png", dpi=300)
plt.show()
```
