```python
import timeit
import random
import matplotlib.pyplot as plt
import math

from good_sorts import quicksort, mergesort, heapsort
from bad_sorts import create_random_list
```

```python
sizes = [math.ceil(1.5**x) for x in range(23)]
max_value = 2**30
runs = 5
```

```python
quick_times = []
merge_times = []
heap_times = []

for n in sizes:
    q_total = m_total = h_total = 0

    for _ in range(runs):
        L = create_random_list(n, max_value)

        start = timeit.default_timer()
        quicksort(L.copy())
        q_total += timeit.default_timer() - start

        start = timeit.default_timer()
        mergesort(L.copy())
        m_total += timeit.default_timer() - start

        start = timeit.default_timer()
        heapsort(L.copy())
        h_total += timeit.default_timer() - start

    quick_times.append(q_total / runs)
    merge_times.append(m_total / runs)
    heap_times.append(h_total / runs)
```

```python
plt.figure()
plt.plot(sizes, quick_times, label="Quick Sort")
plt.plot(sizes, merge_times, label="Merge Sort")
plt.plot(sizes, heap_times, label="Heap Sort")

plt.xlabel("List length")
plt.ylabel("Time (seconds)")
plt.legend()
plt.grid(True)
plt.savefig("experiment4_graph.png", dpi=300)
plt.show()
```
