```python
import timeit
import random
import matplotlib.pyplot as plt
import math

from good_sorts import mergesort, bottom_up_mergesort
from bad_sorts import create_random_list
```

```python
sizes = [math.ceil(1.5**x) for x in range(23)]
max_value = 2**30
runs = 5
```

```python
recursive_times = []
bottomup_times = []

for n in sizes:
    rec_total = 0
    bot_total = 0

    for _ in range(runs):
        L = create_random_list(n, max_value)

        start = timeit.default_timer()
        mergesort(L.copy())
        rec_total += timeit.default_timer() - start

        start = timeit.default_timer()
        bottom_up_mergesort(L.copy())
        bot_total += timeit.default_timer() - start

    recursive_times.append(rec_total / runs)
    bottomup_times.append(bot_total / runs)
```

```python
plt.figure()
plt.plot(sizes, recursive_times, label="Recursive Merge Sort")
plt.plot(sizes, bottomup_times, label="Bottom-Up Merge Sort")

plt.xlabel("List length")
plt.ylabel("Time (seconds)")
plt.legend()
plt.grid(True)
plt.savefig("experiment7_graph.png", dpi=300)
plt.show()
```
