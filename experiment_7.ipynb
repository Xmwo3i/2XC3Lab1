```python
import time
import random
import matplotlib.pyplot as plt

from good_sorts import mergesort
from good_sorts import bottom_up_mergesort
```

```python
def create_random_list(n):
    return [random.randint(0, n) for _ in range(n)]
```

```python
sizes = [500, 1000, 2000, 4000, 8000]
runs = 5

recursive_times = []
bottomup_times = []
```

```python
for n in sizes:
    rec_total = 0
    bot_total = 0

    for _ in range(runs):
        L = create_random_list(n)

        A = L.copy()
        start = time.time()
        mergesort(A)
        rec_total += time.time() - start

        A = L.copy()
        start = time.time()
        bottom_up_mergesort(A)
        bot_total += time.time() - start

    recursive_times.append(rec_total / runs)
    bottomup_times.append(bot_total / runs)
```

```python
plt.figure()
plt.plot(sizes, recursive_times, label="Recursive Merge Sort")
plt.plot(sizes, bottomup_times, label="Bottom-Up Merge Sort")

plt.xlabel("List length")
plt.ylabel("Time (seconds)")
plt.legend()
plt.savefig("experiment7_graph.png", dpi=300)
plt.show()
```
