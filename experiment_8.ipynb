```python
import timeit
import random
import matplotlib.pyplot as plt
import math

from bad_sorts import create_random_list, insertion_sort2
from good_sorts import mergesort, quicksort
```

```python
sizes = [math.ceil(1.5**x) for x in range(23)]
max_value = 2**30
runs = 5
```

```python
insert_times = []
merge_times = []
quick_times = []

for n in sizes:
    ins_total = mer_total = qui_total = 0

    for _ in range(runs):
        L = create_random_list(n, max_value)

        start = timeit.default_timer()
        insertion_sort2(L.copy())
        ins_total += timeit.default_timer() - start

        start = timeit.default_timer()
        mergesort(L.copy())
        mer_total += timeit.default_timer() - start

        start = timeit.default_timer()
        quicksort(L.copy())
        qui_total += timeit.default_timer() - start

    insert_times.append(ins_total / runs)
    merge_times.append(mer_total / runs)
    quick_times.append(qui_total / runs)
```

```python
plt.figure()
plt.plot(sizes, insert_times, label="Insertion Sort")
plt.plot(sizes, merge_times, label="Merge Sort")
plt.plot(sizes, quick_times, label="Quick Sort")

plt.title("Experiment 8: Sorting Algorithm Time Complexity Comparison")
plt.xlabel("List length")
plt.ylabel("Time (seconds)")
plt.legend()
plt.grid(True)
plt.savefig("experiment8_graph.png", dpi=300)
plt.show()
```
