```python
import time
import random
import matplotlib.pyplot as plt

from bad_sorts import insertion_sort2
from good_sorts import mergesort, quicksort
```

```python
def create_random_list(n):
    return [random.randint(0, n) for _ in range(n)]
```

```python
sizes = [5, 10, 20, 30, 40, 50, 75, 100, 150, 200]
runs = 50

insert_times = []
merge_times = []
quick_times = []
```

```python
for n in sizes:
    ins_total = mer_total = qui_total = 0

    for _ in range(runs):
        L = create_random_list(n)

        A = L.copy()
        start = time.time()
        insertion_sort2(A)
        ins_total += time.time() - start

        A = L.copy()
        start = time.time()
        mergesort(A)
        mer_total += time.time() - start

        A = L.copy()
        start = time.time()
        quicksort(A)
        qui_total += time.time() - start

    insert_times.append(ins_total / runs)
    merge_times.append(mer_total / runs)
    quick_times.append(qui_total / runs)
```

```python
plt.figure()
plt.plot(sizes, insert_times, label="Insertion Sort")
plt.plot(sizes, merge_times, label="Merge Sort")
plt.plot(sizes, quick_times, label="Quick Sort")

plt.xlabel("List length")
plt.ylabel("Time (seconds)")
plt.legend()
plt.savefig("experiment8_graph.png", dpi=300)
plt.show()
```
