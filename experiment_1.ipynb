```python
import matplotlib
import random
import timeit
import matplotlib.pyplot as plt
import math
```

```python
from bad_sorts import create_random_list, bubble_sort, selection_sort, insertion_sort
```

```python
lengths = [math.ceil(1.2**x) for x in range(50)]
max_value = 2**30
runs = 10
```

```python
data1, data2, data3 = [], [], []

for n in lengths:
    t1 = t2 = t3 = 0
    
    for _ in range(runs):
        L = create_random_list(n, max_value)

        start = timeit.default_timer()
        bubble_sort(L.copy())
        t1 += timeit.default_timer() - start

        start = timeit.default_timer()
        selection_sort(L.copy())
        t2 += timeit.default_timer() - start

        start = timeit.default_timer()
        insertion_sort(L.copy())
        t3 += timeit.default_timer() - start
    
    data1.append(t1 / runs)
    data2.append(t2 / runs)
    data3.append(t3 / runs)

plt.plot(lengths, data1, label='Bubble Sort')
plt.plot(lengths, data2, label='Selection Sort')
plt.plot(lengths, data3, label='Insertion Sort')
plt.title("Experiment 1: Runtime comparison of sorting algorithms")
plt.xlabel('List Length')
plt.ylabel('Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```
